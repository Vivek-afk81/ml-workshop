### Generate using gpt2

```python
from transformers import AutoModelForCausalLM
import torch
```

```python
gpt2=AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
sentence=" I learn machine learning to enhance the performance"
token_ids=tokenizer(sentence,return_tensors="pt").input_ids

# print(gpt2(token_ids).logits.shape)
outputs=gpt2(token_ids).logits[0, -1]
tokenizer.decode(outputs.argmax())
```

```python
sentence=" I learn machine learning to enhance the"
token_ids=tokenizer(sentence,return_tensors="pt").input_ids
outputs=gpt2(token_ids).logits[0, -1]
final_logits=torch.topk(outputs,7)
print(final_logits)

for index in final_logits.indices:
  print(tokenizer.decode(index))
```
